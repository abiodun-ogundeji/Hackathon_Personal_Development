# Deal with null values

# Views have null values because in our test data the views column is not present so at the time of concat the null view is created so neglect this at that time.

```python
data.isnull().sum().sort_values(ascending=False)
```

```python
def full_nan_by_mean_mod(data):
    nan_columns=data.columns[data.isnull().any()]
    for i in nan_columns:
        if(data[i].dtypes=="object"):
            data[i]=data[i].fillna(data[i].mode()[0])
        else:
            data[i]=data[i].fillna(data[i].mean())
        return data
full_nan_by_mean_mod(data)   
```

```python
##All the missing values is replaced
data.isnull().sum().sort_values(ascending=False)
```
