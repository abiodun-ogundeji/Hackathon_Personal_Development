```python
y=dataset_train.default_status.copy()
dataset_train=X.drop('default_status',1)
dataset_train=dataset_train.drop('Applicant_ID',1)
testId=dataset_test["Applicant_ID"]
dataset_test=dataset_test.drop("Applicant_ID",1)
dataset_train.head()
```

```python
from sklearn import preprocessing

for feat in dataset_train.columns:
    if dataset_train[feat].dtype == 'object':
        y=X.default_status
        lbl = preprocessing.LabelEncoder()
        lbl.fit(np.unique(list(dataset_train[feat].values) + list(dataset_test[feat].values)))
        dataset_train[feat]   = lbl.transform(list(dataset_train[feat].values))
        dataset_test[feat]  = lbl.transform(list(dataset_test[feat].values))

# remove constant columns
remove = []
for col in dataset_train.columns:
    if dataset_train[col].std() == 0:
        remove.append(col)

dataset_train.drop(remove, axis=1, inplace=True)
dataset_test.drop(remove, axis=1, inplace=True)

# remove duplicated columns
remove = []
c = dataset_train.columns
for i in range(len(c)-1):
    v = dataset_train[c[i]].values
    for j in range(i+1,len(c)):
        if np.array_equal(v,dataset_train[c[j]].values):
            remove.append(c[j])

dataset_train.drop(remove, axis=1, inplace=True)
dataset_test.drop(remove, axis=1, inplace=True)


for feat in dataset_train.columns:
    if dataset_train[feat].dtype == 'float64':
        dataset_train[feat][np.isnan(dataset_train[feat])] = dataset_train[feat].mean()
        dataset_test[feat][np.isnan(dataset_test[feat])] = dataset_test[feat].mean()
      
    elif dataset_train[feat].dtype == 'object':
        dataset_train[feat][dataset_train[feat] != dataset_train[feat]] = dataset_train[feat].value_counts().index[0]
        dataset_test[feat][dataset_test[feat] != dataset_test[feat]] = dataset_test[feat].value_counts().index[0]
```
